```python
import nltk
from nltk import FreqDist
from nltk.corpus import stopwords
set(stopwords.words('english'))
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,4)
import os
plt.style.use('dark_background')
```

```python
defoeRaw = open('defoe.txt').read()
defoeWords = nltk.word_tokenize(defoeRaw)
defoe = nltk.Text(defoeWords)
eliotRaw = open('eliot.txt').read()
eliotWords = nltk.word_tokenize(eliotRaw)
eliot = nltk.Text(eliotWords)
stokerRaw = open('stoker.txt').read()
stokerWords = nltk.word_tokenize(stokerRaw)
stoker = nltk.Text(stokerWords)
kerouacRaw = open('kerouac.txt').read()
kerouacWords = nltk.word_tokenize(kerouacRaw)
kerouac = nltk.Text(kerouacWords)
```

```python
alltexts = [defoe, eliot, stoker, kerouac]
```

```python
alltexts
```

```python
stoker.count("time")
```

```python
fdiststoker = FreqDist(stoker)
```

```python
print(fdiststoker)
```

```python
fdiststoker.most_common(50)
```

```python
fdiststoker.most_common(100)
```

```python
stop_words = set(stopwords.words('english'))
filtered_stoker = [w for w in stoker if not w in stop_words]
```

```python
fdistfiltered_stoker = FreqDist(filtered_stoker)
```

```python
fdistfiltered_stoker.most_common(100)
```

```python
def vocab(text):
    return len(text) / len(set(text))
```

```python
vocabularies = {text.name: vocab(text) for text in alltexts}
```

```python
pd.Series(vocabularies)
pd.Series(vocabularies).plot(kind='bar')
```
